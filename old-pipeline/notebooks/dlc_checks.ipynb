```python
import sys
sys.path.append("../")
from utils import metadata as mdutils
from paths import config
from classes.ProcPath import ProcPath
from classes.Experiment import Experiment
import os
import shutil
import deeplabcut
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import argparse
```

```python

cfg = config.M2PConfig()
exps_df = mdutils.get_exps(cfg, primary_only=True, get_excluded=False)
```

```python
# Parameters
jump_threshold = 20  # pixels

# Loop through experiments
completed = True
for index, exp_row in exps_df.iterrows():

    exp_id = exp_row["exp_id"]
    print(f"\nProcessing: {exp_id}")
    
    m2p_paths = ProcPath(cfg=cfg, exp_id=exp_id)
    exp = Experiment(m2p_paths.raw_data_path)

    mov_file = exp.tracking_video.cameras["overhead"].file
    file_name = os.path.split(mov_file)[1]
    (file_name_base, file_ext) = os.path.splitext(file_name)

    proc_mov_path = os.path.join(cfg.video_path, exp_id)
    crop_file = os.path.join(proc_mov_path, file_name_base + "-cropped" + file_ext)

    dlc_path = cfg.dlc_path / "videos"

    print(f"  Processing experiment: {exp_id}"
          f"\n  Video file: {mov_file}"
          f"\n  Cropped video file: {crop_file}"
          f"\n  DLC path: {dlc_path}"
          f"\n  DLC config file: {cfg.dlc_config_path}s"
        )
    
    # Check if video and dlc path exist
    if not os.path.exists(crop_file):
        print(f"  Cropped video not found: {crop_file}")
        break
        
    if not os.path.exists(dlc_path):
        print(f"  DLC path not found: {dlc_path}")
        break

    if not os.path.exists(cfg.dlc_config_path):
        print(f"  DLC config file not found: {cfg.dlc_config_path}")
        break

    print(f"  Video: {crop_file}")
    print(f"  DLC path: {dlc_path}")
    print(f"  DLC config file: {cfg.dlc_config_path}")

    try:
        print("Extract frames with large jumps ...")
        result = deeplabcut.extract_outlier_frames(config=cfg.dlc_config_path, 
                                          videos=[crop_file], 
                                          outlieralgorithm='jump',
                                          epsilon=jump_threshold,
                                          automatic=True,
                                          destfolder=dlc_path)
        
        print(result)
        # if not result:
        #     raise Exception("No result returned.")
        

 
        print("Extract frames with low likelihood ...")
        result = deeplabcut.extract_outlier_frames(config=cfg.dlc_config_path, 
                                          videos=[crop_file],
                                          outlieralgorithm='uncertain',
                                          p_bound=0.01,
                                          automatic=True,
                                          destfolder=dlc_path)

        print(result)
        # if not result:
        #     raise Exception("No result returned.")
        
    except Exception as e:
        print(f"  Error processing {exp_id}: {e}")
        completed = False
        break
    

if completed:
    print("\nDone! Check dlc folders for extracted frames.")
```
